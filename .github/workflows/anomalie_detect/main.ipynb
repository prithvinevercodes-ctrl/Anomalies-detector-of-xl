```python
import pandas as pd
#Load the dataset into a pandas DataFrame
df = pd.read_excel("C:\\Users\\somya\\Downloads\\Mine Dataset.xlsx")
```

```python

#drop the string features
'''df.drop('Power Status', axis = 1,inplace = True)
df.drop('Machine Status', axis = 1, inplace = True)
df.drop('Timestamp', axis = 1, inplace = True)'''
```

```python

# Scale the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python

# Create an Isolation Forest model
clf = IsolationForest(contamination=0.10)  # Contamination is the proportion of outliers in the data
```

```python

# Fit the model and predict anomalies
clf.fit(scaled_data)
predictions = clf.predict(scaled_data)
```

```python

# Anomalies are labeled as -1, normal points are labeled as 1
anomaly_indices = np.where(predictions == -1)[0]
```

```python


for i in df.columns:
    plt.plot(df.index, df[i], c = 'r')
    plt.scatter(anomaly_indices, df[i].loc[anomaly_indices], c = 'b', marker = '*')
    plt.title(i)
    plt.show()
```
